# QUESTION 2: Do some carriers fly further or for longer than others?

### Pseudocode:
#### - Isolate the carrier, flight distance, and flight time columns
#### - Clean the data
#### - Group the data by the carrier and find the mean of the distance and flight time
#### - Sort the values in ascending order to see which carriers have the highest avg flight time and how it trends with flight distance

### Select the data columns we want to look at

```python
c = flights[['carrier','distance','air_time']].dropna()
c
```

### Groupby the carrier to then see the respective distance and air time for each one

```python
d = c.groupby(['carrier'])
d

```

### Take the mean of the flight distance and air time for each carrier

```python
d = d.mean()
```

### Sort the values of each mean for each carrier to find which ones were the highest

```python
d = d.sort_values(by='distance', ascending=False)
d
```

### Single Line

```python
c=flights[["carrier","distance","air_time"]].dropna().groupby("carrier").mean().sort_values(by='distance',ascending=False)
c
```

## FUNCTION CONSTRUCTION

```python
def longest_flights(x):
    c = x[['carrier','distance','air_time']].dropna()
    d = c.groupby(['carrier'])
    d = d.mean()
    d = d.sort_values(by='distance', ascending=False)
    return d
```

```python
longest_flights(flights)
```

# QUESTION 3: Is there a difference in platelets between those who died to heart failure and those who didn't?

### Pseudocode:
#### - Isolate the high blood pressure and death event columns
#### - Group the data by whether they died to heart failure or not
#### - Return the mean among each group for high blood pressure

### Isolating the columns we want to look at (and dropping any possible na values)

```python
heart1 = heart[["high_blood_pressure","DEATH_EVENT"]].dropna()
heart1
```

### Groupby the groups that either did or did not die

```python
heart1 = heart1.groupby(["DEATH_EVENT"])
```

### Take the mean of each group's corresponding boolean of whether they had high blood pressure

```python
heart1.mean()
```

### Single Line

```python
heart1 = heart[["DEATH_EVENT","high_blood_pressure"]].dropna().groupby("DEATH_EVENT").mean()
heart1
```

## FUNCTION CONSTRUCTION

```python
def mean_occurrence(x):
    heart1 = x[['high_blood_pressure','DEATH_EVENT']].dropna()
    a = heart1.groupby(['DEATH_EVENT'])
    b = a.mean()
    return b
```

```python
mean_occurrence(heart)
```
